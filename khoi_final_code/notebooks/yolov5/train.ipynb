```python
import numpy as np
p_r = np.array([[0.57, 0.512, 0.602, 0.544, 0.606], [0.371, 0.332, 0.354, 0.318, 0.419]])
m_ap = np.array([[0.384, 0.344, 0.379, 0.341, 0.456], [0.153, 0.141, 0.171, 0.143, 0.212]])
prec_mean, recall_mean = np.mean(p_r, axis = 1)
prec_std, recall_std = np.std(p_r, axis = 1)
ap05_mean, ap05095_mean = np.mean(m_ap, axis = 1)
ap05_std, ap05095_std = np.std(m_ap, axis = 1)
print(f"mAP0.5: {ap05_mean}+-{ap05_std}, mAp0.5:0.95: {ap05095_mean}:{ap05095_std} Precision: {prec_mean}+-{prec_std}, Recall: {recall_mean}+-{recall_std}")
```

## Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2,2, figsize = (14,14))
axs[0,0].set_xlabel("Inference speed (ms/img)", fontsize = 14)
axs[0,0].set_ylabel("mAP@0.5", fontsize = 14)
axs[0,1].set_xlabel("Inference speed (ms/img)", fontsize = 14)
axs[0,1].set_ylabel("mAP@0.5:0.95", fontsize = 14)
axs[1,0].set_xlabel("Inference speed (ms/img)", fontsize = 14)
axs[1,0].set_ylabel("Precision", fontsize = 14)
axs[1,1].set_xlabel("Inference speed (ms/img)", fontsize = 14)
axs[1,1].set_ylabel("Recall", fontsize = 14)
x = list(range(256, 1536 + 128, 128))
col_names = ['p', 'r', 'mAP0.5', 'mAP0.5:0.95', 'NA_0', 'NA_1', 'NA_2', 'pre-process(ms)', 'inference(ms)', 'NMS/image(ms)']
m_ap, m_ap_interval, p, r = np.zeros((5, 11)), np.zeros((5, 11)), np.zeros((5, 11)), np.zeros((5, 11))

for holdout in range(1,6):
    df = pd.read_csv(f"study_scratch_holdout_{holdout}_best.txt", header = None, sep = ' +', names = col_names, engine = 'python')
    p[holdout - 1, :], r[holdout - 1, :], m_ap[holdout - 1, :], m_ap_interval[holdout - 1, :] = df["p"], df["r"], df["mAP0.5"], df['mAP0.5:0.95']
    axs[0,0].plot(df["inference(ms)"], df["mAP0.5"], label = f"holdout_{holdout}")
    #axs[0,0].axvline(1.5, color = 'k', linestyle = 'dashed', linewidth = 2)
    axs[0,1].plot(df["inference(ms)"], df["mAP0.5:0.95"], label = f"holdout_{holdout}")
    #axs[0,1].axvline(1.5, color = 'k', linestyle = 'dashed', linewidth = 2)
    axs[1,0].plot(df["inference(ms)"], df["p"], label = f"holdout_{holdout}")
    #axs[1,0].axvline(1.5, color = 'k', linestyle = 'dashed', linewidth = 2)
    axs[1,1].plot(df["inference(ms)"], df["r"], label = f"holdout_{holdout}")
    #axs[1,1].axvline(1.5, color = 'k', linestyle = 'dashed', linewidth = 2)
axs[0,0].plot(df["inference(ms)"], m_ap.mean(axis = 0), color = "black", linewidth = 5, label = "Mean")
axs[0,1].plot(df["inference(ms)"], m_ap_interval.mean(axis = 0), color = "black", linewidth = 5, label = "Mean")
axs[1,0].plot(df["inference(ms)"], p.mean(axis = 0), color = "black", linewidth = 5, label = "Mean")
axs[1,1].plot(df["inference(ms)"], r.mean(axis = 0), color = "black", linewidth = 5, label = "Mean")
axs[0,0].scatter([4.313] * 5, [0.1229, 0.1134, 0.1287, 0.1171, 0.1872], label = "Mask-RCNN")
axs[0,1].scatter([4.313] * 5, [0.2880, 0.2803, 0.2889, 0.2833, 0.4141], label = "Mask-RCNN")
axs[0,0].annotate('Better', xy = (3, 0.32), xytext = (5, 0.22), arrowprops = dict(arrowstyle = "fancy", relpos = (0,1), facecolor = 'black'), fontsize = 20)
[axi.legend() for axi in axs.ravel()]
#ax[1,0]
```
